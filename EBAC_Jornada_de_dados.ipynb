#Processamento dos dados

Obtendo todos os dados do Dataset, selecionando apenas as colunas de interesse e os armazenando como itens de uma lista:

```python
data = list()
with open(file="State_of_data_2022.csv", mode="r") as fp:
    header = fp.readline()
    line = fp.readline()
    while line:
        line = line.split(sep=',')
        data.append({"Cargo": line[21], "Experiencia": line[22], "Salário": line[23]})
        line = fp.readline()
len(data)
```

Demonstrando dessa maneira que trabalharemos com as 4271 respostas do questionário.


Estrutura dos dados armazenados:

```python
for linha in data[0:100]:
    print(linha)
```

#Limpeza dos dados

Com uma simples consulta às primeiras 10 linhas do Dataframe criado é fácil notar que há dados inconsistentes, como linhas sem preenchimento ou informações que parecem estar em colunas erradas. Por isso. é necessário fazer primeiro a limpeza dos dados e só então começar as manipulações.

Existem várias abordagens para o tratamento de dados inconsistentes, primeiro, tratando as linhas vazias, será feita a deleção dessas linhas:

```python
#Substitui valores vazios por "NA"
data_table = data_table.replace('', pd.NA)

#Exclui valores "NA" do Dataframe
data_table = data_table.dropna()

data_table.head(15)
data_table.shape
```

Agora tratando as linhas corrompidas (essas podem derivar de diversos fatores: erros de processamento, erros humanos, etc.)

```python
data_table['Salário'] = data_table['Salário'].apply(lambda row: row if 'R$' in row else pd.NA)
```

```python
data_table = data_table.dropna()
```

```python
data_table.head(15)
```

```python
data_table.shape
```

Agora, selecionando as linhas de interesse (no caso dessa análise, contempla somente os que possuíra cargo em Engenharia de dados, Cientista de dados e Analista de dados):

```python
#Demonstrando os valores únicos presentes na coluna 'Cargos'
data_table['Cargo'].unique()
```

```python
positions = ['Analista de Dados/Data Analyst', 'Cientista de Dados/Data Scientist', 'Engenheiro de Dados/Arquiteto de Dados/Data Engineer/Data Architect']
```

```python
data_table['Cargo'] = data_table['Cargo'].apply(lambda row: row if row in positions else pd.NA)
```

```python
data_table = data_table.dropna()
```

```python
data_table.head(n=15)
```

```python
data_table.shape
```

Renomeando as profissões para facilitar as análises:

```python
labels = ['Analista', 'Cientista', 'Engenheira']
labels = dict(zip(positions, labels))
```

```python
data_table['Cargo'] = data_table['Cargo'].apply(lambda row: labels[row])
```

```python
data_table.head(n=15)
```

```python
#Salvando o Dataframe criado com os dados já limpos
data_table.to_csv("data.csv", header=True, index=False)
```

#Agrupamento dos dados
Agrupamento dos dados com uma operação matemática de soma em diversos níveis de agregação.

##Screenshot from 2024-03-20 21-14-29.pngPor salário

```python
data_table['Quantidade'] = 1
```

```python
data_table.head(15)
```

```python
#O Dataframe será agrupado pelo salário e posteriormente agregado pela soma
salario_agg = data_table[['Salário', 'Quantidade']].groupby('Salário').agg('sum').reset_index()
```

```python
salario_agg.head(15)
```

Uma operação comum sobre dados agregados é a ordenação. Como a coluna Salário não é
numericamente ordenável (textos são ordenados por ordem alfabética da primeira letra),
podemos manualmente deanir uma coluna artiacial de ordenamento.

```python
salario_ordem = {
'Menos de R$ 1.000/mês': 0,
'de R$ 1.001/mês a R$ 2.000/mês': 1,
'de R$ 2.001/mês a R$ 3.000/mês': 2,
'de R$ 3.001/mês a R$ 4.000/mês': 3,
'de R$ 4.001/mês a R$ 6.000/mês': 4,
'de R$ 6.001/mês a R$ 8.000/mês': 5,
'de R$ 8.001/mês a R$ 12.000/mês': 6,
'de R$ 12.001/mês a R$ 16.000/mês': 7,
'de R$ 16.001/mês a R$ 20.000/mês': 8,
'de R$ 20.001/mês a R$ 25.000/mês': 9,
'de R$ 25.001/mês a R$ 30.000/mês': 10,
'de R$ 30.001/mês a R$ 40.000/mês': 11,
'Acima de R$ 40.001/mês': 12
}
```

```python
salario_agg['Ordem'] = salario_agg['Salário'].apply(lambda row: salario_ordem[row])
```

```python
salario_agg.head(15)
```

```python
#Ordenando corretamente pela faixa de salário:
salario_agg = salario_agg.sort_values(by='Ordem', ascending=True)
salario_agg.head(15)
```

##Por salário e cargo

```python
salario_cargo_agg = data_table[['Cargo', 'Salário', 'Quantidade']].groupby(['Cargo', 'Salário']).agg('sum').reset_index()
```

```python
salario_cargo_agg['Ordem'] = salario_cargo_agg['Salário'].apply(lambda row: salario_ordem[row])
```

```python
salario_cargo_agg = salario_cargo_agg.sort_values(by=['Cargo', 'Ordem'], ascending=[True, True])
```

```python
salario_cargo_agg.head(100)
```

A informação agregada permite a extração de insights. Por exemplo, vamos analisar o
salário dos analistas de dados.

```python
salario_cargo_agg[salario_cargo_agg['Cargo'] == 'Analista'].head(15)
```

##Por salário, cargo e experiência

```python
salario_cargo_experiencia_agg = data_table[['Cargo', 'Experiencia', 'Salário', 'Quantidade']].groupby(['Cargo', 'Experiencia', 'Salário']).agg('sum').reset_index()
```

```python
salario_cargo_experiencia_agg['Ordem'] = salario_cargo_experiencia_agg['Salário'].apply(lambda row: salario_ordem[row])
```

```python
salario_cargo_experiencia_agg = salario_cargo_experiencia_agg.sort_values(by=['Cargo', 'Ordem', 'Experiencia'], ascending=[True, True, True])
```

```python
salario_cargo_experiencia_agg.head(15)
```

```python
salario_cargo_experiencia_agg[salario_cargo_experiencia_agg['Cargo'] == 'Analista'].head(15)
```

#Visualização dos dados
Utilizando o Seaborn, faremos a visualização dos dados a partir dos Dataframes criados anteriormente.

```python
import seaborn as sns
```

##Por salário

```python
with sns.axes_style('whitegrid'):
    plot = sns.barplot(data=salario_agg, x='Ordem', y='Quantidade', hue='Salário', palette='pastel')
    plot.legend(loc='center left', bbox_to_anchor=(1, 0.5))
```

##Por salário e cargo

```python
with sns.axes_style('whitegrid'):
    plot = sns.barplot(data=salario_cargo_agg, x='Cargo', y='Quantidade', hue='Salário', palette='pastel')
    plot.legend(loc='center left', bbox_to_anchor=(1, 0.5))
```
